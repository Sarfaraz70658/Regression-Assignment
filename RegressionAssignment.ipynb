```python
""" Q-1  What is Simple Linear Regression?

Ans- It is a method used to understand the relationship between two variables: one independent variable (X) and one dependent variable (Y).
 It finds a straight line that best fits the data points, showing how Y changes when X changes.


 Q-2 What are the key assumptions of Simple Linear Regression?

Ans-There is a straight-line relationship between X and Y.
    The errors (differences between predicted and actual values) are random and normally distributed.
    The errors have constant spread (equal variance).
    Each observation is independent of the others.


Q-3 What does the coefficient m represent in the equation Y = mX + c?

Ans-The coefficient m is the slope of the line. It shows how much Y will change when X increases by one unit.


Q-4 What does the intercept c represent in the equation Y = mX + c?

Ans-The intercept c is the point where the line crosses the Y-axis. It shows the value of Y when X is zero.

Q-5 How do we calculate the slope m in Simple Linear Regression?
Ans-The slope is calculated by finding the line that best fits the data.
    It balances how close the line is to all the data points by minimizing the overall error.

Q-6 What is the purpose of the least squares method in Simple Linear Regression?

Ans-The least squares method helps find the best-fitting line by minimizing the sum of squared differences between actual values and predicted values.


Q-7 How is the coefficient of determination (R²) interpreted in Simple Linear Regression?

Ans-R² tells us how well the line explains the variation in the dependent variable (Y).
    If R² = 1 → the line perfectly fits the data.
    If R² = 0 → the line does not explain any variation in Y.

    Values between 0 and 1 show partial fit.

Q-8- What is Multiple Linear Regression?

Ans-It is an extension of simple linear regression where there are two or more independent variables (X1, X2, X3, …) used to predict the dependent variable (Y).


Q-9 What is the main difference between Simple and Multiple Linear Regression?

Ans-Simple Linear Regression: Uses one independent variable.

    Multiple Linear Regression: Uses two or more independent variables.

Q-10 What are the key assumptions of Multiple Linear Regression?

Ans- A linear relationship exists between independent variables and the dependent variable.

     The errors are normally distributed.
     The errors have equal variance (no heteroscedasticity).
    The observations are independent.
    Independent variables are not highly correlated with each other (no strong multicollinearity).


Q-11 What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

Ans-Heteroscedasticity means the spread of errors is not constant across all levels of the independent variables.
    It makes predictions less reliable, standard errors incorrect, and can lead to wrong conclusions in hypothesis testing.


Q-12 How can you improve a Multiple Linear Regression model with high multicollinearity?

Ans-Remove or combine highly correlated variables.
    Use dimensionality reduction methods (like PCA).
    Standardize or center the data.
    Collect more data to reduce dependency.


Q-13- What are some common techniques for transforming categorical variables for use in regression models?

Ans-One-Hot Encoding: Create separate columns for each category (dummy variables).
    Label Encoding: Assign numbers to categories.
    Target/Mean Encoding: Replace categories with average values of the target (used carefully).


Q-14 What is the role of interaction terms in Multiple Linear Regression?

Ans-Interaction terms show how the effect of one variable on the outcome depends on another variable. They help capture combined effects that
    individual variables alone cannot explain.


Q-15 How can the interpretation of intercept differ between Simple and Multiple Linear Regression?

Ans-In Simple Regression, the intercept is the value of Y when X = 0.
    In Multiple Regression, the intercept is the value of Y when all independent variables = 0, which may not always make practical sense.


Q-16 What is the significance of the slope in regression analysis, and how does it affect predictions?

Ans- The slope shows how much the dependent variable changes when the independent variable increases by one unit. A positive slope means Y increases;
     a negative slope means Y decreases.


Q-17 How does the intercept in a regression model provide context for the relationship between variables?

Ans-The intercept anchors the regression line by showing the baseline value of the dependent variable before considering the effects
of the independent variables.


Q-18 What are the limitations of using R² as a sole measure of model performance?

Ans- A high R² does not mean the model is accurate or free of bias.
     It does not show whether the predictors are meaningful.
     It can increase artificially when more variables are added, even if they are irrelevant.


Q-19 How would you interpret a large standard error for a regression coefficient?

Ans-A large standard error means the estimate of that coefficient is unstable and not precise. It suggests the variable may not have a reliable impact
    on the dependent variable.


Q-20 How can heteroscedasticity be identified in residual plots, and why is it important to address it?

Ans- In residual plots, heteroscedasticity appears as a funnel shape (residuals spreading wider or narrower across values). It must be fixed because
     it leads to biased standard errors, making tests and confidence intervals unreliable.


Q-21 What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?

Ans-It means the model might be overfitting. Extra variables are increasing R² artificially, but adjusted R² penalizes irrelevant predictors,
    showing they don’t truly add value.


Q-22 Why is it important to scale variables in Multiple Linear Regression?

Ans- Scaling ensures all variables are on the same scale, preventing large-valued features from dominating, and improving interpretation, stability,
     and model performance.


Q-23 What is polynomial regression?

Ans- Polynomial regression is a type of regression where the relationship between the independent variable and dependent variable is modeled using
     curves instead of a straight line.


Q-24 How does polynomial regression differ from linear regression?

Ans- Linear regression fits a straight line.
     Polynomial regression fits a curve by including higher powers of the independent variable.


Q-25 When is polynomial regression used?

Ans- It is used when data shows a curved (non-linear) relationship that a straight line cannot capture.


Q-26 What is the general equation for polynomial regression?

Ans- It is like linear regression but with higher powers of the input variable (squares, cubes, etc.), which allow the curve to bend.


Q-27 Can polynomial regression be applied to multiple variables?

Ans- Yes, polynomial terms (squares, cross-products) can be added for multiple variables, but the model can become very complex and harder to interpret.


Q-28 What are the limitations of polynomial regression?

Ans- Can overfit easily if too many powers are used.
     Becomes unstable outside the data range (extrapolation).
     Harder to interpret compared to linear regression.


Q-29 What methods can be used to evaluate model fit when selecting the degree of a polynomial?

Ans- Cross-validation.
     Adjusted R².
     Error metrics like Mean Squared Error (MSE).
     Visual inspection of fit versus overfitting.


Q-30 Why is visualization important in polynomial regression?

Ans- Because curves can be complex, visualization helps us see if the chosen degree captures the pattern correctly without overfitting or underfitting. """




```

```python
# Q-31 - How is polynomial regression implemented in Python?

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data (X must be 2D for sklearn)
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([2.5, 3.5, 7.2, 13.5, 25.4, 40.9])  # Non-linear relationship

# Step 1: Transform input features into polynomial features (degree = 2 here)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Step 2: Fit Linear Regression on transformed features
model = LinearRegression()
model.fit(X_poly, y)

# Step 3: Predict values
y_pred = model.predict(X_poly)

# Step 4: Plot results
plt.scatter(X, y, color="blue", label="Actual data")
plt.plot(X, y_pred, color="red", label="Polynomial fit (degree=2)")
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.title("Polynomial Regression Example")
plt.show()

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```
